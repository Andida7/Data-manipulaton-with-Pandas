```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('countries.csv')
```

```python
df
```

```python
# lets compare the population growth for china and america
cntry_grp = df.groupby('country')
```

```python
df['country'].value_counts()['United States']
```

```python
us = cntry_grp.get_group('United States').reset_index(drop=True)
us
```

```python
plt.plot(us.year, us.population/10**9, color = 'b',linestyle = '-', linewidth = 2, label = 'US')
plt.style.use('ggplot')

plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.title('US population growth over 12 years')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
china = cntry_grp.get_group('China').reset_index(drop=True)
china
```

```python
# lets plot a graph for china

plt.plot(china.year, china.population/10**9, color = 'k',linestyle = '--', linewidth = 2, label = 'China')

plt.style.use('ggplot')

plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.legend()

plt.title("China's population growth over 12 years")

plt.tight_layout()
plt.show()
```

```python
# lets draw them together
plt.plot(china.year, china.population/10**9, color = 'k',linestyle = '--', linewidth = 2, label = 'China')
plt.plot(us.year, us.population/10**9, color = 'b',linestyle = '-', linewidth = 2, label = 'US')

plt.style.use('ggplot')

plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.legend()

plt.title("population growth over 12 years")

plt.tight_layout()
plt.show()
```

```python
china.rename(columns={'population':'population_of_china'}, inplace = True)
us.rename(columns={'population':'population_of_US'}, inplace = True)
```

```python
china.drop(columns=['country'], inplace=True)
china
```

```python
us.drop(columns=['country'], inplace=True)
us
```

```python
us['population_of_china'] = china['population_of_china']
```

```python
population = us
population
```

```python
# lets draw them together
plt.plot(population.year
         , population.population_of_china/10**6, color = 'k',linestyle = '--', linewidth = 2, label = 'population_of_china')

plt.plot(population.year, population.population_of_US/10**6, color = 'b',linestyle = '-'
         , linewidth = 2, label = 'population_of_US')

plt.style.use('ggplot')

plt.xlabel('Year')
plt.ylabel('Population in millions')
plt.legend()

plt.title("population growth over 12 years")

plt.tight_layout()
plt.show()
```
