```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# lets drop None values
df.dropna()
```

```python
# dropna() has axis='index' and how='any' by default
# this will drop rows containing none

df.dropna(axis='index', how='any')
```

```python
# but if we change the axsi to 'columns'
# it will drop a column who has None or NaN 
# and how='any' means any column that has atleast one None or NaN

df.dropna(axis='columns', how='any')
```

```python
# but if we change it to how='all' it will check for all a column with all values None or Na
df.dropna(axis='columns', how='all')
```

```python
df.dropna(axis='rows', how='all')
```

```python
# lets try to drop specific columns
# if all

df.dropna(axis='rows', how='all', subset=['email', 'age'])
```

```python
# if any
df.dropna(axis='rows', how='any', subset=['email', 'age'])
```

```python
df
```

```python
df.isna()
```

```python
# the dataframe has another none values that are not written well
# like NA and Missing
# so lets replace them with np.nan

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
# so we can change this none values with 0

df.fillna(0)
```
